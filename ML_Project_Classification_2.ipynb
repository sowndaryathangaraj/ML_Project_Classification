```python
df.dtypes
```

```python
df['diagnosis'].value_counts()
```

```python
x=df.drop(columns=['id','diagnosis','Unnamed: 32'],axis=1)
y=df['diagnosis']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prdctn=model.predict(x_train)
print(x_train_prdctn)
```

```python
accuracy_xtrain=accuracy_score(y_train,x_train_prdctn)
print(accuracy_xtrain)
```

```python
x_test_prdctn=model.predict(x_test)
print(x_test_prdctn)
```

```python
x_tst_accuracy=accuracy_score(y_test,x_test_prdctn)
print(x_tst_accuracy)
```
