```python
le=LabelEncoder()
```

```python
df['Species']=le.fit_transform(df['Species'])
```

```python
x=df.drop(columns=['Id','Species'])
y=df['Species']          
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prdctn=model.predict(x_train)
print(x_train_prdctn)
```

```python
x_tst_prdctn=model.predict(x_test)
print(x_tst_prdctn)
```

```python
x_trn_accuracy=accuracy_score(y_train,x_train_prdctn)
print(x_trn_accuracy)
```

```python
x_tst_accuracy=accuracy_score(y_test,x_tst_prdctn)
print(x_tst_accuracy)
```
